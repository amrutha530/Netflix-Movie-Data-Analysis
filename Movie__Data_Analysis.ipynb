```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import zipfile

with zipfile.ZipFile('/content/archive.zip', 'r') as zip_ref:
    zip_ref.extractall('/content/')

import pandas as pd

df = pd.read_csv('/content/mymoviedb (1).csv', engine='python')
df.head()
df.shape
df.info()
df.isnull().sum()
df.duplicated().sum()
df.describe()
top_movies = df.sort_values(by='Popularity', ascending=False).head(10)

plt.figure(figsize=(10,6))
plt.bar(top_movies['Title'], top_movies['Popularity'])
plt.xticks(rotation=90)
plt.title("Top 10 Popular Movies")
plt.show()
# Convert object columns to numeric
df['Vote_Count'] = pd.to_numeric(df['Vote_Count'], errors='coerce')
df['Vote_Average'] = pd.to_numeric(df['Vote_Average'], errors='coerce')
top_movies = df.sort_values(by='Popularity', ascending=False).head(10)

plt.figure(figsize=(10,6))
plt.bar(top_movies['Title'], top_movies['Popularity'])
plt.xticks(rotation=90)
plt.title("Top 10 Popular Movies")
plt.xlabel("Movie Title")
plt.ylabel("Popularity")
plt.show()
df['Original_Language'].value_counts().head(10).plot(kind='bar')
plt.title("Top 10 Languages")
plt.xlabel("Language")
plt.ylabel("Number of Movies")
plt.show()
plt.figure(figsize=(8,5))
plt.hist(df['Vote_Average'].dropna(), bins=10)
plt.title("Vote Average Distribution")
plt.xlabel("Rating")
plt.ylabel("Number of Movies")
plt.show()
```
